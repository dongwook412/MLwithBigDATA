## Weight

```python
pre_W = np.load('C:\\Users\\Yang\\Desktop\\admm-pruning-master\\weight\\'+'pretrained_'+data_name+'.npy', allow_pickle=True)
admm_W = np.load('C:\\Users\\Yang\\Desktop\\admm-pruning-master\\weight\\'+'admm_'+data_name+'.npy', allow_pickle=True)
re_W = np.load('C:\\Users\\Yang\\Desktop\\admm-pruning-master\\weight\\'+'retrained_'+data_name+'.npy', allow_pickle=True)
```

```python
print(np.sum(pre_W[1]==0.0))
print(np.sum(np.abs(admm_W[1])<=0.1))
print(np.sum(np.abs(re_W[1])<=0.1))
#print(np.sum(sess.run(W_fc2) != 0))
```

```python
# Conv1
k=0
pre_W_ = pre_W[k].reshape(-1)
admm_W_ = admm_W[k].reshape(-1)
re_W_ = re_W[k].reshape(-1)

bins = np.arange(-0.7, 0.7, 0.01)
hist_pre, bins = np.histogram(pre_W_, bins)
hist_admm, bins = np.histogram(admm_W_, bins)
hist_re, bins = np.histogram(re_W_, bins)

plt.figure(figsize=(20, 7))
plt.subplot(1,3,1)
plt.hist(pre_W_, bins)
plt.xlabel('Weight value')
plt.ylabel('Count')
plt.subplot(1,3,2)
plt.hist(admm_W_, bins)
plt.xlabel('Weight value')
plt.ylabel('Count')
plt.subplot(1,3,3)
plt.hist(re_W_, bins)
plt.xlabel('Weight value')
plt.ylabel('Count')

plt.show()
```
